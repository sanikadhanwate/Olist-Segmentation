```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import MiniBatchKMeans
from sklearn.metrics import silhouette_score


df = pd.read_csv(r"C:\Users\Niyati\Downloads\olist_cleaned_dataset\olist_cleaned_dataset.csv")   # <-- change if needed


num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
features = num_cols 

X = df[features].copy()
X = X.replace([np.inf, -np.inf], np.nan)
X = X.fillna(X.median(numeric_only=True))

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=min(20, X_scaled.shape[1]), random_state=42)
X_pca = pca.fit_transform(X_scaled)

sample_n = min(5000, X_pca.shape[0])
idx = np.random.choice(X_pca.shape[0], sample_n, replace=False)
X_sample = X_pca[idx]

best_k, best_sil = None, -1
for k in range(3, 11):        # range of clusters to test
    km = MiniBatchKMeans(n_clusters=k, random_state=42, batch_size=2048, n_init=10)
    km.fit(X_sample)
    labels = km.predict(X_sample)
    sil = silhouette_score(X_sample, labels)
    if sil > best_sil:
        best_k, best_sil = k, sil

print(f" Best k = {best_k} | Silhouette = {best_sil:.4f}")

kmeans = MiniBatchKMeans(n_clusters=best_k, random_state=42, batch_size=2048, n_init=10)
df["kmeans_cluster"] = kmeans.fit_predict(X_pca)

summary_cols = [
    'recency_days','num_orders','total_spent','avg_order_value',
    'avg_items_per_order','avg_item_price_per_order','avg_review',
    'avg_installments','num_product_categories','latitude','longitude'
]

cluster_summary = df.groupby("kmeans_cluster")[summary_cols].mean().round(2)
print("\n====== Cluster Profile ======")
print(cluster_summary)

cluster_summary.to_csv("kmeans_cluster_profile.csv", index=False)

#2D PCA PLOT
pca2 = PCA(n_components=2, random_state=42)
X_2d = pca2.fit_transform(X_scaled)

plt.figure(figsize=(8,5))
plt.scatter(X_2d[:,0], X_2d[:,1], s=5, c=df["kmeans_cluster"])
plt.title(f"K-Means Clusters (k={best_k}) — 2D PCA")
plt.xlabel("PC1"); plt.ylabel("PC2")
plt.show()


id_col = "customer_unique_id" if "customer_unique_id" in df.columns else None

labels_out = pd.DataFrame({
    id_col if id_col else "row_id": df[id_col] if id_col else np.arange(len(df)),
    "kmeans_cluster": df["kmeans_cluster"]
})
```

```python
from sklearn.cluster import KMeans
inertia = []
K = range(2, 12)

for k in K:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_pca)   # use PCA features you created earlier
    inertia.append(km.inertia_)

plt.figure(figsize=(7,4))
plt.plot(K, inertia, marker='o')
plt.title("Elbow Plot")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia")
plt.grid(alpha=0.3)
plt.show()
```

```python
df_original = df.copy()   # save copy BEFORE scaling/PCA
df_original["kmeans_cluster"] = df["kmeans_cluster"]
metrics = ["recency_days", "num_orders", "total_spent", "avg_order_value",
           "avg_items_per_order", "avg_review"]

cluster_means = df_original.groupby("kmeans_cluster")[metrics].mean()

cluster_means.T.plot(kind='bar', figsize=(10,6))
plt.title("Cluster Profiles (Original Scale)")
plt.ylabel("Mean Value")
plt.xlabel("Customer Metrics")
plt.legend(title="Cluster")
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

neighbors = NearestNeighbors(n_neighbors=10)
neighbors_fit = neighbors.fit(X_pca)
distances, indices = neighbors_fit.kneighbors(X_pca)

# Sort distances (largest first helps detect elbow)
distances = np.sort(distances[:, 9])
plt.figure(figsize=(8,4))
plt.plot(distances)
plt.title("K-Distance Graph for DBSCAN (use elbow for eps)")
plt.xlabel("Points sorted by distance")
plt.ylabel("10-NN Distance")
plt.show()
```

```python
eps_value = 2.5

dbscan = DBSCAN(eps=eps_value, min_samples=20, n_jobs=-1)
db_labels = dbscan.fit_predict(X_pca)
print("Clusters found:", len(set(db_labels)) - (1 if -1 in db_labels else 0))
print("Noise points:", sum(db_labels == -1))
```

```python
plt.figure(figsize=(7,5))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=db_labels, cmap="viridis", s=5)
plt.title("DBSCAN Clusters on PCA space")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
